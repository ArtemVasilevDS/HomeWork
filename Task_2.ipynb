```python
from bs4 import BeautifulSoup
import requests
```

```python
req = requests.get("http://www.vedomosti.ru/business/articles/2016/04/17/637993-vstrecha")
```

```python
import bs4
parser = bs4.BeautifulSoup(req.text, 'lxml')
for script in parser(["script", "style"]):
    script.extract()
x = parser.find('div', attrs = {'class':'b-document__body b-social__layout-mutation'})    
x = x.text
    
lines = (line.strip() for line in x.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
x = '\n'.join(chunk for chunk in chunks if chunk)

title = parser.find('h1', attrs = {'class': 'title'})


```
