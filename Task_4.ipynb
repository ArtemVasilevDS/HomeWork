```python
import math
import numpy as np
from matplotlib import pylab as pl

%matplotlib inline
```

```python
x = np.linspace(0, 3, 10)

def f1(x):
    return x

def exp_(x):
    return map(lambda x: math.exp(x), x)

def quadric_(x):
    return map(lambda x: x**2, x)

plt.plot(x, f1(x), x, exp_(x), x, quadric_(x))
plt.xlabel('$x$ $value$', fontsize = 14)
plt.ylabel('$Functions$ $value$', fontsize = 14)
plt.text(2.2, 3.8, r'$f(x) = x$', fontsize = 14)
plt.text(2.2, 8, r'$f(x) = x^2$', fontsize = 14)
plt.text(2.2, 17, r'$f(x) = e^x$',fontsize = 14)
plt.title(r'$Pyplot$ $tutorial$', fontsize = 14)
plt.savefig('plot.pdf')
plt.savefig('plot.png')
plt.show()
```

```python
def func(x, y):
    return 3*x*y + x - 2*y

X = np.linspace(0, 5, 9)
Y = np.linspace(0, 5, 9)
Z = 3*X*Y + X - 2*Y
data = np.zeros((9,9))

for i, x in list(enumerate(X)):
    for j, y in list(enumerate(Y)):
        data[i, j] = func(x,y)
    
```

```python
fig = plt.figure()
s = fig.add_subplot(1, 1, 1)
im = s.imshow(data, extent=(X[0], X[-1], Y[0], Y[-1]), origin='lower')
plt.title('$f(x,y) = 3xy + x - 2y$', fontsize = 16)
plt.xlabel('$x$', fontsize = 16)
plt.ylabel('$y$', fontsize = 16)

fig.colorbar(im)
```
